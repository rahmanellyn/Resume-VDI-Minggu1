```python

from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df= pd.read_csv('diabetes_null.csv')
print(df.head(10))

```

```python
df.info()
```

```python

# Plot 1: Distribusi Kadar Glukosa Berdasarkan Outcome (0: Tidak Diabetes, 1: Diabetes)
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Glucose', hue='Outcome', kde=True, bins=30, palette='coolwarm')
plt.title('Distribusi Kadar Glukosa Berdasarkan Outcome Diabetes')
plt.xlabel('Kadar Glukosa')
plt.ylabel('Frekuensi')
plt.legend(title='Outcome', labels=['Tidak Diabetes', 'Diabetes'])
plt.show()

# Plot 2: Scatter plot BMI vs Age berdasarkan Outcome
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Age', y='BMI', hue='Outcome', palette='coolwarm', alpha=0.7)
plt.title('Hubungan Antara Usia, BMI, dan Outcome Diabetes')
plt.xlabel('Usia')
plt.ylabel('BMI')
plt.legend(title='Outcome', labels=['Tidak Diabetes', 'Diabetes'])
plt.show()
```
