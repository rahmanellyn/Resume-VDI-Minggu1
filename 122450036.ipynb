Visualisasi data adalah bidang yang berkekbang pesat dengan banyak sekali hasil penelitian batu dan sistem baru yang dikembangkan baru-baru ini. Penelitian dan praktisi dari berbagai bidang telah dikontribusi pada keberhasilan luar biasa visualisasi data, yang didorong oleh sebagian besar(jika tidak semua) domain dan aplikasi.

Visualisasi Data adalah pengubahan data menjadi data  fisik yang menarik yang dapat dibaca dengan mudah oleh manusia. Dari data Abstrak menjadi data yang memiliki panjang, poisi,bentuk,warna dan lainnya. 
Visualisasi data sangat cocok untuk diterapkan ppada data yang besar dan mempermudah untuk meng interpresentasikan hasil analisis data kepada data science.

Visualisasi data berulang memiliki alur sebagai berikut:
1. Import data 
import data adalah mengambil data yang diperlukan dari yang diinginkan sumber data
2. Persiapan data
Persiapan data adalah mempersiapkan data yang diimpor visualisasi, misalnya dengan menormalkan nilai, mengoreksi entri yang salah, dan interpolasi nilai yang hilang.
3. Manipulasi data
yaitu memilih data yang akan divisualisasikan (alis pemfilteran dari kimunitas ) dan mungkin dengan operasi umum lainnya seperti bergabung dan pengelompokan.
4. Pemetaan
Memetakan data yang dipeoleh dari atas proses keprimitif geometris
5. Renderung
Rendering adalah mengubah data geometri diatas menjadi representasi visual.
